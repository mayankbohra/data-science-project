```python
# Power Method to Find Largest Eigen Value and Eigen Vector

import numpy as np

a = np.array([[0,2,0,1,3,2,0,2],[0,0,2,0,2,1,0,0],[1,1,0,0,2,1,0,0],[0,3,1,0,0,2,1,0],[2,0,2,1,0,0,0,3],[3,1,1,2,0,0,1,2],[0,2,0,3,0,1,0,1],[0,2,2,0,0,0,1,0]],dtype = 'float')
        

#Xo.....
Xo = np.array([[1],[1],[1],[1],[1],[1],[1],[1]])


#Where to end the iterations...
maXo_iteration = int(input("Enter where to end..."))

step = 1
while True:
   #Scalar prod
    Xo = np.matmul(a,Xo)
    
    # Finding new Eigen value and Eigen vector
    lambdaNew = max(abs(Xo))
    
    Xo = Xo/lambdaNew
    
    # Displaying Eigen value and Eigen Vector
    print('\nstep',(step))                  # The Step
    
    print('Eigen Value is  = ',(lambdaNew))    #
    print('Eigen Vector: is ')
    for i in range(8):
        print((Xo[i]))
    
    # Checking Max iteration
    step = step + 1
    if step > maXo_iteration:
        break
```
