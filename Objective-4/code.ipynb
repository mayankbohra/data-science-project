```python
import matplotlib.pyplot as plt
import numpy as np

total = [2,5,7,3,2,2,1,1]

money_range = ['0-300','300-600','600-900','900-1200','1200-1500','1500-1800','18000-2100','2100-2400']

indices = np.arange(len(money_range))
# width = 0.4

plt.bar(indices, total, color = 'b')
plt.xticks(ticks=indices, labels = money_range, rotation='vertical')

plt.xlabel('Amount of Money Spent on R&D(Total) [in crores]')
plt.ylabel('Frequency of States')
plt.title('Frequency of states having total amount of money spent on R&D in a particular range')

plt.show()
```
